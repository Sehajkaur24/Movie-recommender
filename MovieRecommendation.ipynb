Loading Dataset from filesystem

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/content/ml-latest-small/ratings.csv")
```

```python
df.head()
```

Since we are build content based movie recommender system, therefore, we don't need timestamp

```python
df = df.drop('timestamp', axis=1)
```

```python
df.head()
```

loading movie titles

```python
movies = pd.read_csv("/content/ml-latest-small/movies.csv")
```

```python
movies.head()
```

we don't need genres for now, so lets drop it

```python
movies = movies.drop('genres', axis=1)
```

```python
movies.head()
```

let us combine movies with ratings dataset

```python
ratings = pd.merge(df, movies, on='movieId')
```

```python
ratings.head()
```

Getting average and count of rating of each title

```python
title_ratings = ratings.groupby('title')['rating'].mean().to_frame()
title_ratings['numRating'] = ratings.groupby('title')['rating'].count()
title_ratings.head()
```

Creating using **User-Movie Matrix** using pandas pivot_table

```python
moviemat = ratings.pivot_table(index='userId', columns='title', values='rating')
moviemat.head()
```

Lets check recommendations for **"Toy Story (1995)"**

```python
corr_toy_story = moviemat.corrwith(moviemat['Toy Story (1995)']).to_frame('correlation')
corr_toy_story = corr_toy_story.join(title_ratings.numRating)
```

```python
corr_toy_story[corr_toy_story.numRating > 90].sort_values('correlation', ascending=False).head()
```

thus, movies to be watched after **"Toy Story (1995)"** are :-


1.   Toy Story 2 (1999)
2.   Incredibles, The (2004)
3.   Finding Nemo (2003)
4.   Aladdin (1992)

### Building model using above dataset

```python
class MovieRecommender:
  moviemat = moviemat
  title_ratings = title_ratings

  def getMovies(self, movieName):
    corr_table = moviemat.corrwith(self.moviemat[movieName]).to_frame('correlation')
    corr_table = corr_table.join(self.title_ratings.numRating)
    return corr_table[corr_table.numRating > 90].sort_values('correlation', ascending=False).head().index
```

```python
rec = MovieRecommender()
[ i for i in rec.getMovies('Aladdin (1992)')]
```
